```python
#연습문제3
def calc(c,*args):
    if c=='+':
        sum1=0
        for i in args:
            sum1+=i
        print(sum1)
    if c=='-':
        sum11=args[0]*2
        for i in args:
            sum11-=i
        print(sum11)
    if c=='*':
        sum2=1
        for i in args:
            sum2*=i
        print(sum2)

calc('+',1,2,3,4,5)
calc('-',1,2,3,4,5)
calc('*',1,2,3,4,5)
```

```python
base_url='https://search.naver.com/search.naver'
params={
    'where':'news',
    'query':'무역전쟁'
}
news_list=[]
result_list=[]
for i in range(3):
    resp=requests.get(base_url,params=params)
    soup=BeautifulSoup(resp.text)
    news_list_items=soup.select('ul.type01>li')
     
    for news_item in news_list_items:
        anchor_title = news_item.find('a',class_='_sp_each_title')
        
        desc=news_item.find_all('dd')[1]
        title=anchor_title.text
        url=anchor_title.get('href')
        
        #request 보내기
        news_detail_resp = requests.get(url,verify=False)
        content=news_detail_resp.text
        
        result_list.append({
            'title':title,
            'desc':desc,
            'url':url,
            'contnet':content
        })
```

```python
def replace_iframe(soup,base_url):
    iframe_tag=soup.find_all('iframe')
    
    for iframe in iframe_tag:
        iframe_src=iframe.get('src')
        
        #urljoin
        iframe_src=urljoin(base_url,iframe_src)
        
        if iframe_src!=base_url:
            resp=requests.get(iframe_src)
            new_soup=BeautifulSoup(resp.text)
            
            new_soup=replace_iframe(new_soup,resp.url)
            iframe.replace_with(new_soup)
    return soup
```

```python
url = 'https://finance.naver.com/sise/sise_index_day.nhn'

target_codes = ['KOSPI', 'KOSDAQ']
max_page = 3

result_dict = {}

for code in target_codes:
    result_dict[code] = []
    for page in range(1, max_page+1):
        params = {
            'code': code,
            'page': page
        }
        resp = requests.get(url, params=params)
        soup=BeautifulSoup(resp.text)
        tr_tags = soup.find_all('tr')
        
        for tr in tr_tags:
            date = tr.find('td', class_='date')
            if not date:
                continue
            date = date.text
            
            number = tr.find('td', class_='number_1').text
        
            result_dict[code].append(dict(date=date, number=number))
```

```python
result_dict
```

```python
kospi_df=pd.DataFrame(result_dict['KOSPI'])
kosdaq_df=pd.DataFrame(result_dict['KOSDAQ'])
```

```python
kospi_df['number'] = kospi_df['number'].apply(lambda x: "".join(x.split(',')))
kospi_df['number'] = kospi_df['number'].astype(np.float64)
kospi_df['date'] = kospi_df['date'].astype(np.datetime64)
```

```python
kosdaq_df['number'] = kosdaq_df['number'].apply(lambda x: "".join(x.split(',')))
kosdaq_df['number'] = kosdaq_df['number'].astype(np.float64)
kosdaq_df['date'] = kosdaq_df['date'].astype(np.datetime64)
```

```python
kospi_df = kospi_df.sort_values(by='date')
kosdaq_df = kosdaq_df.sort_values(by='date')
```

```python
plt.figure(figsize=(20, 8))
plt.plot(kospi_df['date'], kospi_df['number'])
plt.ylabel('value',fontsize=20)
plt.xlabel('date',fontsize=20)
plt.show()
```
