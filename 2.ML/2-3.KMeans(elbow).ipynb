```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

X,y=make_blobs(n_samples=15000,
               n_features=2,
               centers=20,
               cluster_std=0.5,
               shuffle=True,
               random_state=0)

distortions=[]

for i in range(2,10):
    km=KMeans(n_clusters=i,init='random',n_init=100,max_iter=300,tol=1e-04,random_state=0)
    km=km.fit(X)
    distortions.append(km.inertia_)

plt.plot(range(2,10),distortions)
plt.show()
```
